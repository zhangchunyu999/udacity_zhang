### Reflection

### 1. Describe your pipeline. As part of the description, explain how you modified the draw_lines() function.

My pipeline consisted of 5 steps. 
First, I converted the images to grayscale with the function of 'cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)'
![alt text][image1]

Second, I converted the images to edges with Canny Edge Detection.I used the function of 'cv2.Canny()'
![alt text][image2]

Third, in order to pull only the lane lines out of the image, I masked out a region of interest in the image with the function of 'cv2.fillPoly()'
![alt text][image3]

Fourth, I implemented the Hough Transform on edge detected images with the function of 'cv2.HoughLinesP()', then drew the line on the images.In order to draw a single line on the left and right lanes, I modified the draw_lines() function,    
![alt text][image4]

Finally, I copy the line onto the original image to make my current pipeline as the following image.
